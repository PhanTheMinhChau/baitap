```python
#Nhập các thư viện cần thiết
%matplotlib inline
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import altair as alt
```

```python
#Đọc dữ liệu
mb_df = pd.read_csv("microbiome.csv")
mb_df.head()
```

```python
mb_df.iloc[3]
```

Vẽ Box plot

```python
mb_df['Tissue'].plot(kind='box')
```

Bài tập: Hãy cải tiến biểu đồ box plot trên để đẹp hơn

```python
fig, ax = plt.subplots(figsize=(7,16))
ax.set_facecolor('#FFC0CB')
sns.boxplot(data=mb_df[['Tissue']],color="#00FF00",width=0.2, fliersize=20, flierprops = dict(marker='$❤$',markerfacecolor='r'))
```

VẼ SCATTER PLOT VỚI 2 TRỤC LÀ TISSUE - STOOL

```python
mb_df_2 = mb_df[['Tissue', 'Stool']]
mb_df_2.head()
```

```python
mb_df_2.plot.scatter(x = 'Tissue', y = 'Stool',figsize=(12,6))
```

Bài tập: Hãy chỉnh sửa để biểu đồ đẹp hơn

```python
data=mb_df_2.loc[[0] , ['Tissue', 'Stool'] ]
print(data)
```

```python
#code của bạn ở đây
fig, ax = plt.subplots(figsize=(15,9))
ax.set_facecolor('#FFFF66')
sns.scatterplot(data=mb_df_2,x = 'Tissue', y = 'Stool', color="#FF0000",marker='X',s=380)
```

VẼ HISTOGRAM

```python
mb_df['Stool'].plot(kind='hist')
```

```python
mb_df['Stool'].plot(kind='hist', cumulative= True)
```

```python
mb_df['Stool'].plot(kind='hist', cumulative= True, bins = 50)
```

Bài tập: Hãy làm đẹp biểu đồ trên

```python
fig = plt.figure()
ax = fig.add_axes([1, 1, 1.5, 1.5])
left, width = .25, .5
bottom, height = .25, .5
right = left + width
top = bottom + height
sns.histplot(data=mb_df['Stool'],color='#00FF00',edgecolor='none',element="poly")
sns.despine()
ax.text(0.5*(left+right), 0.5*(bottom+top), 'Stool', horizontalalignment='center',verticalalignment='center',fontsize=50, color='red',transform=ax.transAxes)
sns.set_context("talk")
```

```python
fig = plt.figure()
ax = fig.add_axes([1, 1, 1.5, 1.5])
left, width = .25, .5
bottom, height = .25, .5
right = left + width
top = bottom + height
sns.histplot(data=mb_df['Stool'],color='#00FF00', cumulative= True,edgecolor='none',element="poly")
sns.despine()
ax.text(0.5*(left+right), 0.5*(bottom+top), 'Stool', horizontalalignment='center',verticalalignment='center',fontsize=50, color='red',transform=ax.transAxes)
sns.set_context("talk")
```
